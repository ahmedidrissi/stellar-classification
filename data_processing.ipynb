## **Importing Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## **2. Data Cleaning**

```python
# checking missing values
df.isnull().sum()
```

```python
# dropping unnecessary columns
df.drop(['obj_ID','run_ID','rerun_ID','cam_col','field_ID','fiber_ID'], axis = 1, inplace=True)
```

```python
# mapping numbers to class types
df["class"]=df["class"].map({"GALAXY":0,"STAR":1,"QSO":2})
```

```python
df.head(10)
```

```python
df.describe().T
```

## **3. Feature Engineering**

```python
# finding correlations
corr_matrix = df.corr()
plt.figure(figsize=(10,10))
sns.heatmap(corr_matrix,cmap = 'Blues',annot=True)
```

```python
corr_matrix['class'].sort_values(ascending=False)
```

## **4. Data Analysis and Data Visualization**

```python
plt.pie(df['class'].value_counts(),autopct="%1.1f%%",labels=['GALAXY','STAR','QSO'])
plt.legend()
```

```python
plt.figure(figsize=(10,10))
sns.countplot(x=df['class']) # plot Type column
```
